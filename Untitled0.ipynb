```python
# 🧠 STEP 2: Import Required Libraries
import torch
import numpy as np
import cv2
import io
from PIL import Image
from google.colab import files
import open_clip
import face_alignment
from torchvision import transforms
import mediapipe as mp

device = "cuda" if torch.cuda.is_available() else "cpu"

# Load CLIP
model, _, preprocess_clip = open_clip.create_model_and_transforms(
    'ViT-B-32', pretrained='laion2b_s34b_b79k'
)
model = model.to(device).eval()

# Load Face Alignment Model
fa = face_alignment.FaceAlignment(face_alignment.LandmarksType.TWO_D, device=device)

# Load MediaPipe for hair/beard segmentation
mp_selfie_segmentation = mp.solutions.selfie_segmentation
seg_model = mp_selfie_segmentation.SelfieSegmentation(model_selection=1)

# 🧼 STEP 3: Hair/Beard Removal Helper
def remove_hair_and_beard(image: np.ndarray) -> np.ndarray:
    rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    results = seg_model.process(rgb_image)
    mask = results.segmentation_mask

    binary_mask = (mask < 0.4).astype(np.uint8) * 255
    kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))
    binary_mask = cv2.morphologyEx(binary_mask, cv2.MORPH_OPEN, kernel)

    inpainted = cv2.inpaint(image, binary_mask, inpaintRadius=3, flags=cv2.INPAINT_TELEA)
    return inpainted
```

```python
# 🧱 STEP 4: Full Preprocessing Function
def preprocess_face_pipeline(pil_img: Image.Image, fa, size=224) -> Image.Image:
    img_np = np.array(pil_img.convert("RGB"))
    landmarks_list = fa.get_landmarks(img_np)

    if not landmarks_list:
        raise ValueError("No face detected.")

    landmarks = landmarks_list[0]
    x1, y1 = landmarks[:, 0].min(), landmarks[:, 1].min()
    x2, y2 = landmarks[:, 0].max(), landmarks[:, 1].max()
    cx, cy = (x1 + x2) / 2, (y1 + y2) / 2
    size_crop = max(x2 - x1, y2 - y1) * 1.4

    x1 = int(cx - size_crop / 2)
    y1 = int(cy - size_crop / 2)
    x2 = int(cx + size_crop / 2)
    y2 = int(cy + size_crop / 2)

    cropped = img_np[max(0, y1):y2, max(0, x1):x2]
    cropped = cv2.resize(cropped, (size, size))

    no_hair = remove_hair_and_beard(cropped)

    gray = cv2.cvtColor(no_hair, cv2.COLOR_BGR2GRAY)
    gray_rgb = cv2.cvtColor(gray, cv2.COLOR_GRAY2RGB)

    return Image.fromarray(gray_rgb)
```

```python
# 🧠 STEP 5: Upload Attractive Faces and Build Attractiveness Vector
print("Upload attractive male face images.")
uploaded = files.upload()

embeddings = []

for name, data in uploaded.items():
    try:
        print(f"Processing: {name}")
        img = Image.open(io.BytesIO(data))
        processed = preprocess_face_pipeline(img, fa)
        display(processed)

        tensor = preprocess_clip(processed).unsqueeze(0).to(device)

        with torch.no_grad():
            emb = model.encode_image(tensor)
            emb = emb / emb.norm(dim=-1, keepdim=True)
            embeddings.append(emb.squeeze(0))

        print(f"✅ Successfully processed: {name}")
    except Exception as e:
        print(f"❌ Failed: {name} — {type(e).__name__}: {e}")
```

```python
# ✅ Step 6: Create and Save the Attractiveness Vector
if embeddings:
    attractiveness_vector = torch.stack(embeddings).mean(dim=0)
    attractiveness_vector = attractiveness_vector / attractiveness_vector.norm()

    # Save it for future use
    torch.save(attractiveness_vector, "attractiveness_vector.pt")
    print("📁 Attractiveness vector saved as 'attractiveness_vector.pt'")
else:
    raise RuntimeError("❌ No valid attractive faces processed.")
```

```python
if not embeddings:
    raise RuntimeError("No valid attractive faces processed.")

# Compute the attractiveness anchor vector
attractiveness_vector = torch.stack(embeddings).mean(dim=0)
attractiveness_vector /= attractiveness_vector.norm()
print("\n✅ Attractiveness vector created.")

# 🧪 STEP 6: Upload Test Face and Score
print("\nUpload a test face.")
test_file = files.upload()
test_name, test_data = next(iter(test_file.items()))
test_img = Image.open(io.BytesIO(test_data))

try:
    processed_test = preprocess_face_pipeline(test_img, fa)
    display(processed_test)
    tensor_test = preprocess_clip(processed_test).unsqueeze(0).to(device)

    with torch.no_grad():
        test_emb = model.encode_image(tensor_test)
        test_emb /= test_emb.norm(dim=-1, keepdim=True)
        similarity = (test_emb @ attractiveness_vector.T).item()

    print(f"\n📷 Test Face: {test_name}")
    print(f"🔥 Attractiveness similarity score: {similarity:.4f} (range: -1 to 1)")
except Exception as e:
    print(f"❌ Error processing test face: {e}")
```
